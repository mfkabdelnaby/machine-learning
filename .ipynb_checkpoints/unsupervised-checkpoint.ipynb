## Clustering

### K-means clustering
* Finds clusters of samples
* Number of clusters must be specified
* k-means remembers the mean of each cluster ("centroids"), and assigns new sample to its nearest centroid



* **`KMeans()`**
    * **`n_clusters`**

````python
# import our varialbes
from variables import points, new_points

points = np.array(points)
new_points = np.array(new_points)

xs = points[:,0]
ys = points[:,1]

plt.scatter(xs,ys)
plt.show()
````

The scatter plot suggests that there are 3 distinct clusters

````python
# Import KMeans
from sklearn.cluster import KMeans

# Create a KMeans instance with 3 clusters: model
model = KMeans(n_clusters=3)

# Fit model to points
model.fit(points)

# Determine the cluster labels of new_points: labels
labels = model.predict(new_points)

# Print cluster labels of new_points
print(labels)
````

````python
# Import pyplot
import matplotlib.pyplot as plt

# Assign the columns of new_points: xs and ys
xs = new_points[:,0]
ys = new_points[:,1]

# Make a scatter plot of xs and ys, using labels to define the colors
plt.scatter(xs,ys,c=labels,alpha=0.5)

# Assign the cluster centers: centroids
centroids = model.cluster_centers_

# Assign the columns of centroids: centroids_x, centroids_y
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]

# Make a scatter plot of centroids_x and centroids_y
plt.scatter(centroids_x,centroids_y,marker='D',s=50)
plt.show()
````

The clustering looks great!

### Evaluating a clustering

**Inertia** measures how far the samples are from centroid of its cluster. The lower the better

**`KMeans().inertia_`**

````python
samples_df = pd.read_csv('datasets\Grains\seeds.csv')
samples_df.head()
````

````python
# import the samples
samples_df = pd.read_csv('datasets\Grains\seeds.csv')
samples = samples_df.values #array

ks = range(1, 6)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(samples)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
````

The inertia decreases very slowly from 3 clusters to 4, so it looks like 3 clusters would be a good choice for this data.

Our data frame, `samples`, contains some labels for the clusters in the last column; the labels are 1,2, and 3. We can use them as a measure of cluster's accuracy. This will be done using **Cross Tabulation**

````python
# last column
varieties = samples_df.iloc[:,7].tolist()
print(varieties)
````

````python
# Create a KMeans model with 3 clusters: model
model = KMeans(n_clusters=3)

# Use fit_predict to fit model and obtain cluster labels: labels
labels = model.fit_predict(samples)

# Create a DataFrame with labels and varieties as columns: df
df = pd.DataFrame({'labels': labels, 'varieties': varieties})

# Create crosstab: ct
ct = pd.crosstab(df['labels'],df['varieties'])

# Display ct
print(ct)
````

The crosstab above shows the three labels are separated really well accross the 3 clusters

## Transforming features for better clusterings

### Standarizations
Sometimes when some clusters have high variances compared to others, it becomes harder for the ML model to distinguish between them. Thus, it is adviced to perform **Standarization** to unify the variances to 1, and means to 0 for all clusters.

**NOTE**:
* Use **`fit()/transform()`** with **StandardScaler**
* Use **`fit()/predict()`** with **KMeans**

This standarization and clustering steps will be done using pipelines. Notice that they are done differenlty here than in the `supervised-learning` notebook. But after all they are functioning exactly the same.

````python
# Perform the necessary imports
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


fish_df = pd.read_csv('datasets/fish.csv')

samples = fish_df.drop('Bream',axis=1).values
species = fish_df['Bream'].values
````

````python
kmeans = KMeans(n_clusters=4)
labels = kmeans.fit_predict(samples)

df = pd.DataFrame({'labels':labels,'species':species})
ct = pd.crosstab(df['labels'],df['species'])
print(ct)
````

Looks clustering was not performed perfectly here. Let's study the effect of standarizarion where all values will be standarized so that they have the same scale

````python
# Create scaler: scaler
scaler = StandardScaler()

# Create KMeans instance: kmeans
kmeans = KMeans(n_clusters=4)

# Create pipeline: pipeline
pipeline = make_pipeline(scaler,kmeans)

# Fit the pipeline to samples
pipeline.fit(samples)

# Calculate the cluster labels: labels
labels = pipeline.predict(samples)

# Create a DataFrame with labels and species as columns: df
df = pd.DataFrame({'labels':labels,'species':species})

# Create crosstab: ct
ct = pd.crosstab(df['labels'],df['species'])

# Display ct
print(ct)
````

It looks like the fish data separates really well into 4 clusters!

### Single vs complete linkage Hierarchical clustering
* **`linkage(data_samples,method='complete')`** it is a measure of distance between clusters
    * **`method = 'complete'`** means the distance between clusters is max of distance between their samples
    * **`method = 'single'`** distance between clusters is the distance of closest points in both clusters

In the next example you will study the stocks data in which each column corresponds to daily dollar price movement in one trading day, and each raw corresponds to a company.

Notice that the standarization method used here is **Normalizer** which scales each point independently. Unlike **StandardScaler** which unifies variance and removes means

````python
movements_df = pd.read_csv('datasets\stocks.csv')
movements_df.head()
````

````python
movements = movements_df.drop('Unnamed: 0',axis=1).values #array
companies = movements_df['Unnamed: 0'].tolist()
````

````python
# Import Normalizer
from sklearn.preprocessing import Normalizer

# Create a normalizer: normalizer
normalizer = Normalizer()

# Create a KMeans model with 10 clusters: kmeans
kmeans = KMeans(n_clusters=10)

# Make a pipeline chaining normalizer and kmeans: pipeline
pipeline = make_pipeline(normalizer,kmeans)

# Fit pipeline to the daily price movements
pipeline.fit(movements)
````

````python
# Import pandas
import pandas as pd

# Predict the cluster labels: labels
labels = pipeline.predict(movements)

# Create a DataFrame aligning labels and companies: df
df = pd.DataFrame({'labels': labels, 'companies': companies})

# Display df sorted by cluster label
print(df.sort_values('labels'))
````

````python
# Perform the necessary imports
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Import normalize
from sklearn.preprocessing import normalize

# Normalize the movements: normalized_movements
normalized_movements = normalize(movements)

# Calculate the linkage: mergings
mergings = linkage(normalized_movements,method='complete')

# Plot the dendrogram
dendrogram(mergings,labels=companies,leaf_rotation=90,leaf_font_size=6)
plt.show()
````

SciPy hierarchical clustering doesn't fit into a sklearn pipeline, so you'll need to use the `normalize()` function from `sklearn.preprocessing` instead of `Normalizer`.

Let's try performing `single` linkage hierarchical clustering. Will it produce a different dendrogram??

````python
# Calculate the linkage: mergings
mergings_single = linkage(normalized_movements,method='single')

# Plot the dendrogram
dendrogram(mergings_single,labels=companies,leaf_rotation=90,leaf_font_size=6)
plt.show()
````

## Dimension reduction

* More efficient storage and computation
* Remove less-informative "noise" features
* Easier classification, regression, ...


### Principal Component Analysis

Will use the **decorrelation** type
* Rotates data samples to be aligned with axes
* Shift samples so they have mean 0
* Principal components = direction of variance.



* **`PCA`** is sklearn component which is exactly like **StandardScaler**
    * **`fit()`** learns transformation from given data (doesnt apply them)
    * **`transform()`** applies the learned transformation, and could be applied to new data
    
**IMPORTANT NOTE** features of the data could be correlated. However, features of the PCA are NOT correlated.

````python
grains = pd.read_csv('datasets\Grains\seeds-width-vs-length.csv').values
````

````python
# Perform the necessary imports
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Assign the 0th column of grains: width
width = grains[:,0]

# Assign the 1st column of grains: length
length = grains[:,1]

# Scatter plot width vs length
plt.scatter(width, length)
plt.axis('equal')
plt.show()

# Calculate the Pearson correlation
correlation, pvalue = pearsonr(width,length)

# Display the correlation
print(correlation)
````

As you would expect, the width and length of the grain samples are highly correlated.

````python
# Import PCA
from sklearn.decomposition import PCA

# Create PCA instance: model
model = PCA()

# Apply the fit_transform method of model to grains: pca_features
pca_features = model.fit_transform(grains)

# Assign 0th column of pca_features: xs
xs = pca_features[:,0]

# Assign 1st column of pca_features: ys
ys = pca_features[:,1]

# Scatter plot xs vs ys
plt.scatter(xs, ys)
plt.axis('equal')
plt.show()

# Calculate the Pearson correlation of xs and ys
correlation, pvalue = pearsonr(xs, ys)

# Display the correlation
print(correlation)
````

### Direction of the First Principal Component

The first principal component of the data is the direction in which the data varies the most. 

* **`PCA().components_`**

````python
# Make a scatter plot of the untransformed points (two features)
plt.scatter(grains[:,0], grains[:,1])

# Create a PCA instance: model
model = PCA()

# Fit model to points
model.fit(grains)

# Get the mean of the grain samples: mean
mean = model.mean_

# Get the first principal component: first_pc
first_pc = model.components_[0,:]

# Plot first_pc as an arrow, starting at mean
plt.arrow(mean[0], mean[1], first_pc[0], first_pc[1], color='black', width=0.03)

# Keep axes on same scale
plt.axis('equal')
plt.show()
````

### Intrinsic dimension

**Intrinsic dimension** the number of features required to approximate the dataset. It is also defined as the number of PCA featrues that have significant variance.

* **`pca.explained_variance_`** variances of each p.component
* **`pca.n_components_`** the number of p.component

**NOTE** you need to standarize the data first before studying the intrinsic dimension

````python
fish_df = pd.read_csv('datasets/fish.csv')

samples = fish_df.drop('Bream',axis=1).values
species = fish_df['Bream'].values
````

````python
# Perform the necessary imports
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

# Create scaler: scaler
scaler = StandardScaler()

# Create a PCA instance: pca
pca = PCA()

# Create pipeline: pipeline
pipeline = make_pipeline(scaler,pca)

# Fit the pipeline to 'samples'
pipeline.fit(samples)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.show()
````

Since PCA features 0 and 1 have significant variance, the intrinsic dimension of this dataset appears to be **2**.

````python
# Create a PCA model with 2 components: pca
pca = PCA(n_components=2)

# Fit the pipeline to 'samples'
pipeline = make_pipeline(scaler,pca)

# Fit the PCA instance to the scaled samples
pipeline.fit(samples)

# Transform the scaled samples: pca_features
pca_features = pca.transform(samples)

# Print the shape of pca_features
print(pca_features.shape)
````

You've successfully reduced the dimensionality from 6 to 2

### Clustering Text data using TruncatedSVD

* **`TruncatedSVD`** perform PCA on sparse arrays in csr_matrix format, such as word-frequency arrays

In this exercise you will perfrom text dimention reduction on Wikipedia articles data. The word frequency array using `tf-idf` is already provided. All you need is to create the pipeline of TruncatedSVD and KMeans to perfrom word clustering

````python
# it is word frequency array
df = pd.read_csv('datasets\Wikipedia articles\wikipedia-vectors.csv', index_col=0)
df.head()
````

````python
import pandas as pd
from scipy.sparse import csr_matrix

df = pd.read_csv('datasets\Wikipedia articles\wikipedia-vectors.csv', index_col=0)
articles = csr_matrix(df.transpose())
titles = list(df.columns)

print(articles)
print(titles)
````

````python
# Perform the necessary imports
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline

# Create a TruncatedSVD instance: svd
svd = TruncatedSVD(n_components=50)

# Create a KMeans instance: kmeans
kmeans = KMeans(n_clusters=6)

# Create a pipeline: pipeline
pipeline = make_pipeline(svd,kmeans)

# Fit the pipeline to articles
pipeline.fit(articles)

# Calculate the cluster labels: labels
labels = pipeline.predict(articles)

# Create a DataFrame aligning labels and titles: df
df = pd.DataFrame({'label': labels, 'article': titles})

# Display df sorted by cluster label
print(df.sort_values('label'))
````

## Non-negative matrix factorization (NME)

* Dimension reduction technique
* NME models are *interpretable* (unlike PCA)
* It can only applied to non-negative features
* A sample can be reconstructed by multiplying components by feature values and add up (matrix multiplication)

**Ex1** use `NMF` with Wikipedia articles which is a frequency array using tf-idf

````python
# Import NMF
from sklearn.decomposition import NMF

# Create an NMF instance: model
model = NMF(n_components=6)

# Fit the model to articles
model.fit(articles)

# Transform the articles: nmf_features
nmf_features = model.transform(articles)

# Print the NMF features
print(nmf_features[:10])
````

````python
# Create a pandas DataFrame: df
df = pd.DataFrame(nmf_features,index=titles)

# Print the row for 'Anne Hathaway' article
print(df.loc['Anne Hathaway'])

# Print the row for 'Denzel Washington'
print(df.loc['Denzel Washington'])
````

Notice the NMF feature 3 has the highest value. This means that both articles are reconstructed using mainly the 3rd NMF component.

````python
# Import pandas
import pandas as pd

# Create a DataFrame: components_df
components_df = pd.DataFrame(model.components_)

# Select row 3: component (the 4th component (which is a topic/cluster in this example))
component = components_df.iloc[3]

# Print result of nlargest, which returns top 5 frequent words
print(component.nlargest())
````

* column `4707` corresonds to 'film'
* 983 = 'award'
* 11273 = 'starred'
* 10260 = 'role'
* 155 = 'actress'

Therefore, you can simply determine the topic of the cluster from its most frequent words.


**Ex2: Use NMF on LED image dataset**

In the previous dataset, each row was an article, and each row is a feature (word). Similarly, in the image dataset below, each row is an image, and each column is a pixel (feature). Each row represents a 13x8 image. You will see how NMF learns parts of the images. You will use 7 components for the NMF as the number of LED cells are 7

````python
samples = pd.read_csv('datasets/LED.csv').values
samples
````

````python
# Import pyplot
from matplotlib import pyplot as plt

# Select the 0th row: digit
digit = samples[0]

# Print digit
print(digit)

# Reshape digit to a 13x8 array: bitmap
bitmap = digit.reshape(13,8)

# Print bitmap
print(bitmap)

# Use plt.imshow to display bitmap
plt.imshow(bitmap, cmap='gray', interpolation='nearest')
plt.colorbar()
plt.show()
````

In the next exercise you will see how NMF **learns** parts of images

````python
# define a function that displays image by 1D array
def show_as_image(sample):
    bitmap = sample.reshape((13, 8))
    plt.figure()
    plt.imshow(bitmap, cmap='gray', interpolation='nearest')
    plt.colorbar()
    plt.show()
    
# Import NMF
from sklearn.decomposition import NMF

# Create an NMF model: model
model = NMF(n_components=7)

# Apply fit_transform to samples: features
features = model.fit_transform(samples)

# Call show_as_image on each component
for component in model.components_:
    show_as_image(component)

# Assign the 0th row of features: digit_features
digit_features = features[0]

# Print digit_features
print(digit_features)
````

Notice how each component is a part of the image. This is the main feature of NMF which, that it **learns** from clusters of the dataset, unlike PCA

````python
# Import PCA
from sklearn.decomposition import PCA

# Create a PCA instance: model
model = PCA(n_components=7)

# Apply fit_transform to samples: features
features = model.fit_transform(samples)

# Call show_as_image on each component
for component in model.components_:
    show_as_image(component)
````

 Notice that the components of PCA do not represent meaningful parts of images of LED digits!

### Building recommender system using NMF

We can use the great advantage of NMF that their features describe topics from the word-frerquency array, and build recommendation system that recommends other topics with similar NMF feature values.

We can measure the similarity between two NMF features using **Cosine similarity** which uses the angle between the two feature lines. Higher values means more similar. The max value is 1 when angle = 0 degree.

* **`.dot()`** a dataframe method which computes the cosine similarity between rows of df and the article (instance) of interest

**Ex3: Which articles are similar to 'Cristiano Ronaldo'?**

````python
# NMF features abtained earlier
print(nmf_features[:10])
````

````python
# Perform the necessary imports
import pandas as pd
from sklearn.preprocessing import normalize

# Normalize the NMF features: norm_features
norm_features = normalize(nmf_features)

# Create a DataFrame: df
df = pd.DataFrame(norm_features,index=titles)

# Select the row corresponding to 'Cristiano Ronaldo': article
article = df.loc['Cristiano Ronaldo']

# Compute the dot products: similarities
similarities = df.dot(article)

# Display those with the largest cosine similarity
print(similarities.nlargest(10))
````
